````python
def mas_at_pos(a_pos,data:str,verbose=False):
    import re
    lines = data.strip().splitlines()
    line_count = len(lines)
    line_len = len(lines[0])
    def mas_matches(m):
        for (offset,pat) in m:
            pos = (a_pos[0] + offset[0],a_pos[1]+offset[1])
            if pos[0] < 0 or pos[1] < 0 or pos[0] >= line_len or pos[1] >= line_count: return False
            if verbose: print(f"checking if data[{pos=}] which is {lines[pos[1]][pos[0]]} matches {pat}")
            if re.match(pat,lines[pos[1]][pos[0]]) is None: return False
        return True
    return any([mas_matches(m) for m in mas])
````
